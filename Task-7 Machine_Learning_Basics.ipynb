```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load your cleaned dataset from previous days
df = pd.read_csv("task-5_data_cleaning.csv")
df = df.drop_duplicates()
df['Attendance_Pct'] = df['Attendance_Pct'].fillna(df['Attendance_Pct'].mean())
df['Midterm_Score'] = df['Midterm_Score'].fillna(df['Midterm_Score'].mean())

print("Features available for training:", df.columns.tolist())
```

**Cell 2: Feature Selection & Train-Test Split**

```python
# Select feature (X) and target (Y)
X = df[['Midterm_Score']]
y = df['Final_Score']

# Perform Train-Test Split (Using a small test size due to dataset constraints)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

print(f"Training set size: {X_train.shape[0]} samples")
print(f"Testing set size: {X_test.shape[0]} samples")
```

**Cell 3: Train the Linear Regression Model Python**

```python
# Initialize the model
model = LinearRegression()

# Train (fit) the model on training data
model.fit(X_train, y_train)

print("Model training complete!")
print(f"Intercept (Beta_0): {model.intercept_:.2f}")
print(f"Coefficient (Beta_1): {model.coef_[0]:.2f}")
```

**Cell 4: Evaluate Model Performance**

```python
# Generate predictions on the unseen test data
y_pred = model.predict(X_test)

# Calculate metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("--- Model Performance Metrics ---")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R2) Score: {r2:.2f}")
```
