**5.Functions**

```python
# Defining a progress calculator function
def calculate_progress(days_done, total_days=30):
    """Calculates completion progress percentage."""
    percentage = (days_done / total_days) * 100
    return round(percentage, 2)

# Calling the function for Day 2
progress_percent = calculate_progress(days_done=2)

print("--- Function Output ---")
print(f"My Internship Progress: {progress_percent}%")
```

```python
# Program 1: Internship Resource Budget Planner

def analyze_budget(budget_usd, exchange_rate=83.50):
    # Convert budget to local currency (INR)
    budget_inr = budget_usd * exchange_rate
    print(f"Total Budget: ${budget_usd} USD (~₹{budget_inr:,.2f} INR)")
    print("-" * 40)
    
    # List of learning resources and their costs in INR
    resources = {
        "Python Advanced Book": 1500,
        "Premium ML Course": 4500,
        "Cloud Computing Subscription": 8000,
        "AI Research Paper Database Access": 3000
    }
    
    spent_so_far = 0
    
    # Loop through resources and calculate if we can afford them
    for resource, cost in resources.items():
        if spent_so_far + cost <= budget_inr:
            spent_so_far += cost
            print(f"✅ Purchased: {resource} (Cost: ₹{cost})")
        else:
            print(f"❌ Skipped: {resource} (Cost: ₹{cost}) - Over Budget!")
            
    remaining_budget = budget_inr - spent_so_far
    print("-" * 40)
    print(f"Total Spent: ₹{spent_so_far:,.2f} INR")
    print(f"Remaining Budget: ₹{remaining_budget:,.2f} INR")

# Run the budget planner with a $150 USD budget limit
analyze_budget(budget_usd=150)
```

```python
# Program 2: Student Performance Grade Analyzer

def calculate_grade(scores_list):
    if not scores_list:
        return "No scores provided."
    
    # Calculate sum using a loop
    total_score = 0
    for score in scores_list:
        total_score += score
        
    # Calculate average
    average = total_score / len(scores_list)
    
    # Determine Grade using conditional statements
    if average >= 90:
        grade = "A (Outstanding)"
    elif average >= 80:
        grade = "B (Excellent)"
    elif average >= 70:
        grade = "C (Good)"
    elif average >= 50:
        grade = "D (Pass)"
    else:
        grade = "F (Fail)"
        
    return round(average, 2), grade

# Test scores for 5 assignments
my_scores = [85, 92, 78, 90, 88]

# Run the analyzer
avg_score, final_grade = calculate_grade(my_scores)

print("--- Academic Report Card ---")
print("All Assignment Scores:", my_scores)
print(f"Average Score Percentage: {avg_score}%")
print(f"Final Academic Grade: {final_grade}")
```
