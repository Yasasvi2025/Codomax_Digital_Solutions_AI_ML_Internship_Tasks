**Cell 1: Generate the Dirty Dataset**

```python
import csv

# Create dirty data with duplicates and missing values
dirty_data = [
    ["Student_ID", "Name", "Department", "Attendance_Pct", "Midterm_Score", "Final_Score"],
    ["STU_101", "Aarav Sharma", "Computer Science", 92.5, 78, 85],
    ["STU_102", "Diya Patel", "Data Science", 88.0, 92, 90],
    ["STU_103", "Vihaan Gupta", "Computer Science", None, 65, 72], # Missing attendance
    ["STU_104", "Ananya Reddy", "Artificial Intelligence", 95.0, 88, 94],
    ["STU_102", "Diya Patel", "Data Science", 88.0, 92, 90],        # Duplicate Row
    ["STU_105", "Kabir Verma", "Data Science", 81.5, None, 58],   # Missing midterm score
    ["STU_106", "Sai Joshi", "Artificial Intelligence", 89.8, 84, 82],
    ["STU_104", "Ananya Reddy", "Artificial Intelligence", 95.0, 88, 94] # Duplicate Row
]

# Write to CSV
with open("task-5_data_cleaning.csv", mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerows(dirty_data)

print("Dirty dataset 'task-5_data_cleaning.csv' generated successfully!")
```

**Cell 2: Load and Check for Duplicates & Missing Values**

```python
import pandas as pd

# Load dirty dataset
df = pd.read_csv("task-5_data_cleaning.csv")
print("Initial Dataset Shape:", df.shape)

# Check for missing values per column
print("\n--- Missing Value Count ---")
print(df.isnull().sum())

# Check for total duplicate rows
print("\nTotal Duplicate Rows:", df.duplicated().sum())
```

**Cell 3: Data Cleaning (Drop Duplicates & Impute Missing Values)**

```python
# 1. Remove duplicate rows permanently
df_clean = df.drop_duplicates().copy()

# 2. Handle missing values using column means
df_clean['Attendance_Pct'] = df_clean['Attendance_Pct'].fillna(df_clean['Attendance_Pct'].mean())
df_clean['Midterm_Score'] = df_clean['Midterm_Score'].fillna(df_clean['Midterm_Score'].mean())

print("Cleaned Dataset Shape:", df_clean.shape)
```

**Cell 4: Understand Dataset Statistics**

```python
print("--- Summary Statistics for Numerical Columns ---")
print(df_clean.describe())
```
