**Cell 1: Load Resources and Reconstruct Model Pipelines**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Load our curated student records dataset
df = pd.read_csv("task-5_data_cleaning.csv")
df = df.drop_duplicates()
df['Attendance_Pct'] = df['Attendance_Pct'].fillna(df['Attendance_Pct'].mean())
df['Midterm_Score'] = df['Midterm_Score'].fillna(df['Midterm_Score'].mean())

# Define variables using the true independent feature in our dataset 
# (Note: Midterm_Score acts as the metric predictor for Final_Score)
X = df[['Midterm_Score']]
y = df['Final_Score']

# Split data using fixed random state for structural consistency
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# Instantiate and fit the model estimator
regressor = LinearRegression()
regressor.fit(X_train, y_train)
print("Linear Regression model successfully re-instantiated! ✓")
```

**Cell 2: Run Corrected Target Predictions & Custom Inference**

```python
# 1. Generate predictions on the formal evaluation test matrix split
y_pred = regressor.predict(X_test)

print("--- Test Partition Prediction Matrix ---")
for actual, predicted in zip(y_test, y_pred):
    print(f"Actual Final Score: {actual} | Predicted Final Score: {predicted:.2f}")

print("\n--- Custom Feature Inference Prediction ---")
# 2. Map standard inference vector (e.g., scoring an 85.0 on the Midterm)
custom_input = pd.DataFrame([[85.0]], columns=['Midterm_Score']) # Fixed feature name tracking warning
custom_pred = regressor.predict(custom_input)
print(f"A student with an input score of 85.0 is predicted to achieve a Final Score of: {custom_pred[0]:.2f}")
```
