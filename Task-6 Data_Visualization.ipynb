**Cell 1: Import Libraries and Load Data**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load our cleaned data from Day 5
# (Ensure the file path matches where your clean dataframe is saved)
df = pd.read_csv("task-5_data_cleaning.csv")
# Quick drop of duplicates if using raw file directly
df = df.drop_duplicates()
df['Attendance_Pct'] = df['Attendance_Pct'].fillna(df['Attendance_Pct'].mean())
df['Midterm_Score'] = df['Midterm_Score'].fillna(df['Midterm_Score'].mean())

print("Data loaded successfully for visualization! Total records:", len(df))
```

**Cell 2: Scatter Plot (Attendance vs. Final Score)**

```python
plt.figure(figsize=(7, 5))
plt.scatter(df['Attendance_Pct'], df['Final_Score'], color='blue', alpha=0.7, edgecolors='black', s=100)

plt.title('Scatter Plot: Attendance Pct vs. Final Score', fontsize=12, fontweight='bold', pad=15)
plt.xlabel('Attendance Percentage (%)', fontsize=10)
plt.ylabel('Final Score', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.savefig('scatter_attendance_vs_final.png', dpi=300)
plt.show()
```

**Cell 3: Bar Chart (Average Final Score by Department)**

```python
# Calculate department averages
dept_avg = df.groupby('Department')['Final_Score'].mean().reset_index()

plt.figure(figsize=(8, 5))
plt.bar(dept_avg['Department'], dept_avg['Final_Score'], color=['#2B6CB0', '#319795', '#4A5568'], edgecolor='black', width=0.5)

plt.title('Bar Chart: Average Final Score by Department', fontsize=12, fontweight='bold', pad=15)
plt.xlabel('Department', fontsize=10)
plt.ylabel('Mean Final Score', fontsize=10)
plt.ylim(0, 100) # Scale appropriately for score ranges
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.savefig('bar_department_avg_score.png', dpi=300)
plt.show()
```

**Cell 4: Line Chart (Progress Trend: Midterm vs. Final Score)**

```python
plt.figure(figsize=(8, 5))
plt.plot(df['Name'], df['Midterm_Score'], marker='o', color='#E53E3E', linewidth=2, label='Midterm Score')
plt.plot(df['Name'], df['Final_Score'], marker='s', color='#2B6CB0', linewidth=2, linestyle='--', label='Final Score')

plt.title('Line Chart: Progress Trend (Midterm vs. Final Score)', fontsize=12, fontweight='bold', pad=15)
plt.xlabel('Student Name', fontsize=10)
plt.ylabel('Scores', fontsize=10)
plt.xticks(rotation=15)
plt.legend(loc='lower right')
plt.grid(True, linestyle='--', alpha=0.3)

plt.tight_layout()
plt.savefig('line_score_trend.png', dpi=300)
plt.show()
```
