**Cell 1: Pipeline Initialization and Data Preparation**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Load our curated student records
df = pd.read_csv("task-5_data_cleaning.csv")
df = df.drop_duplicates()
df['Attendance_Pct'] = df['Attendance_Pct'].fillna(df['Attendance_Pct'].mean())
df['Midterm_Score'] = df['Midterm_Score'].fillna(df['Midterm_Score'].mean())

# Define target tracking dimensions (X = Feature matrix, y = Target vector)
X = df[['Midterm_Score']]
y = df['Final_Score']

# Split data using fixed state for reproducibility
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

print("--- Data Matrices Formed ---")
print(f"X_train Shape: {X_train.shape} | y_train Shape: {y_train.shape}")
```

**Cell 2: Model Training Routine**

```python
# Instantiate the Linear Regression model estimator
regressor = LinearRegression()

print("Starting model training cycle...")
# Train the estimator weights using structural features
regressor.fit(X_train, y_train)
print("Model training completed successfully! ✓")

# Output the mathematical parameter vectors derived from training
print(f"Learned Intercept (Constant Bias): {regressor.intercept_:.4f}")
print(f"Learned Slope Coefficient (Feature Weight): {regressor.coef_[0]:.4f}")
```
